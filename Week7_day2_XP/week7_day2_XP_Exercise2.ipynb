```python
import pandas as pd

# Load the CSV file into a pandas DataFrame
df = pd.read_csv('/content/archive3/clean_dataset.csv')

# Select the columns of interest
columns_of_interest = ['Age', 'Debt', 'YearsEmployed', 'Income']

# Calculate statistics
statistics = {
    'Mean': df[columns_of_interest].mean(),
    'Median': df[columns_of_interest].median(),
    'Mode': df[columns_of_interest].mode().iloc[0]  # mode() returns a DataFrame, so we take the first row
}

# Convert statistics dictionary to a DataFrame for better readability
statistics_df = pd.DataFrame(statistics)

# Display the statistics
print(statistics_df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a pandas DataFrame
df = pd.read_csv('/content/archive3/clean_dataset.csv')

# Plot Age vs. Debt
plt.plot(df['Age'], df['Debt'], 'o')
plt.xlabel('Age')
plt.ylabel('Debt')
plt.title('Age vs. Debt')
plt.show()
```
